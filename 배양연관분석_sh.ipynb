```python
import pandas as pd

file_path = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/배양검사결과.csv'
data = pd.read_csv(file_path)

columns_to_keep = ['VTHOS_NO', 'RGNO', 'ANTBT_NM', 'RSLT_CONT']
data = data[columns_to_keep]

data = data.dropna()

# 'RSLT_CONT' 열에서 'NEG', 'POS', 'I', 'D', 'ETC', 'NULL' 값을 가진 행 제외 #감수성과 내성 외 제외
filtered_data = data[~data['RSLT_CONT'].isin(['NEG', 'POS', 'I', 'D', 'ETC', 'NULL'])]
display(filtered_data)

output_file_path = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/배양1_sh.csv'
filtered_data.to_csv(output_file_path, index=False)
```

```python
import pandas as pd

file_path = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/배양1_sh.csv'
data = pd.read_csv(file_path)

# ANTBT_NM 열에 해당 조건을 만족하는 행 제거 #분석의 초점, 중복성 방지, 특정 연구 질문 위해 특정 항생제 제거
unwanted_antbt = ['ESBL', 'β-lactamase', 'Inducible Clindarmycin Resistance', 'Cefocxitin Screen']
data = data[~data['ANTBT_NM'].isin(unwanted_antbt)]
display(data)

output_file_path = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/배양2_sh.csv'
data.to_csv(output_file_path, index=False)
```

```python
import pandas as pd

csv_file_path = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/배양2_sh.csv'

data = pd.read_csv(csv_file_path)

data = data[['ANTBT_NM', 'RSLT_CONT']]

#pivot 테이블 생성 (약물 이름을 행, 결과를 열로)
pivot_table = data.pivot_table(data, index='ANTBT_NM', columns='RSLT_CONT', aggfunc=len, fill_value=0)

print(pivot_table)
```

```python
output_csv_file_path = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/배양3_sh.csv'
pivot_table.to_csv(output_csv_file_path)
```
