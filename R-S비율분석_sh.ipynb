```python
import pandas as pd

output_csv_file = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/처리된_데이터_RS2.csv'
output_filtered_csv_file = '/content/drive/MyDrive/챗봇/데이터-20250331T081748Z-001/데이터/배양 전처리/상위_40개_GERM_NM_데이터_sh.csv'

data = pd.read_csv(output_csv_file)

germ_nm_counts = data['GERM_NM'].value_counts()

top_40_germs = germ_nm_counts.head(40).index.tolist()

filtered_data = data[data['GERM_NM'].isin(top_40_germs)]

filtered_data.to_csv(output_filtered_csv_file, index=False)

print("상위 40개 GERM_NM 데이터 종류를 포함한 행을 저장한 CSV 파일 생성 완료.")
```
