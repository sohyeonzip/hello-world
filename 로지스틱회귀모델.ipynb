## 5. 텐서플로에서 실행하는 로지스틱 회귀 모델

### 1.환경 준비

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. 데이터 준비

```python
# x1 = np.array([2, 4, 6, 8, 10, 12, 14])       # 공부시간
# x2 = np.array([3, 5, 2, 4, 3, 5, 6])          # 출석일수
# x3 = np.array([0, 1, 0, 2, 3, 4, 4])          # 모의시험 응시 횟수
# y = np.array([0, 0, 0, 1, 1, 1, 1])           # 합/불합 여부

x1 = np.array([2, 4, 6, 18, 20, 22, 34])       # 공부시간
x2 = np.array([3, 5, 2, 34, 33, 45, 56])          # 출석일수
x3 = np.array([0, 1, 0, 12, 13, 14, 14])          # 모의시험 응시 횟수
y = np.array([0, 0, 0, 1, 1, 1, 1])           # 합/불합 여부
```

```python
X = np.column_stack((x1, x2, x3))
X
```

```python
#X = pd.DataFrame([x1, x2, x3]).T
#X
```

### 3. 모델 실행

```python
model = Sequential()
model.add(Dense(1, input_dim=3, activation='sigmoid'))

# 교차 엔트로피 오차 함수를 이용하기 위하여 'binary_crossentropy'로 설정합니다.
model.compile(optimizer='sgd' ,loss='binary_crossentropy')                      # opti 에서 sgd의 gd는 그레디언트. 이진분류는 묻따 바이너리_크로스엔트로피.

model.fit(X, y, epochs=200)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=3, activation='relu'))
#model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 교차 엔트로피 오차 함수를 이용하기 위하여 'binary_crossentropy'로 설정합니다.
model.compile(optimizer='adam' ,loss='binary_crossentropy', metrics=['accuracy'])                      # opti 에서 sgd의 gd는 그레디언트. 이진분류는 묻따 바이너리_크로스엔트로피.
history=model.fit(X, y, epochs=200, batch_size=2)
```

### 4. 모델 테스트

```python
#임의의 학습시간을 집어넣어 합격 예상 확률을 예측해 보겠습니다.

hour = 7
attend = 5
test = 1
prediction = model.predict(np.array([[hour, attend, test]]))

print("%.f시간을 공부할 경우, 합격 예상 확률은 %.01f%%입니다" % (hour, prediction * 100))
```
